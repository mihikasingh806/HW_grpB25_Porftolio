```python
import pandas as pd
import numpy as np
stock_rets = pd.read_excel('spx_returns_weekly.xlsx', sheet_name='s&p500 rets', index_col = 'date')
factor_rets = pd.read_excel('factor_pricing_data_weekly.xlsx', sheet_name='factors (excess returns)', index_col = 'Date')

start_date = pd.Timestamp(2015, 1, 3)
stock_rets = stock_rets.loc[start_date:]


end_date = stock_rets.index[-1] + pd.Timedelta(days = 1)
factor_rets = factor_rets.loc[start_date:end_date]
risk_free_rets = pd.read_excel('factor_pricing_data_weekly.xlsx', sheet_name='risk-free rate', index_col = 'Date')[start_date:end_date]

display(risk_free_rets)
# display(factor_rets)
display(stock_rets)
excess_stocks_rets = stock_rets.sub(risk_free_rets['RF'], axis = 0)

display(excess_stocks_rets)
```

```python
import statsmodels.api as sm

FREQ = 52
tickers = excess_stocks_rets.columns
betas = []
annualized_means = []
for ticker in tickers:
    y = excess_stocks_rets[ticker]
    X = sm.add_constant(factor_rets[['MKT']])
    model = sm.OLS(y, X).fit()
    alpha, beta = model.params
    betas.append(beta)
    annualized_means.append(y.mean()*FREQ)

df = pd.DataFrame({'mkt beta': betas, 'mean': annualized_means}, index = tickers)
```

```python
# 1.2
annualized_abs_alpha = {}
LFPM_types = ['CAPM', 'FF', '4F']
factor_sets = [
    ['MKT'],
    ['MKT', 'SMB', 'HML'],
    ['MKT', 'HML', 'RMW', 'UMD'],
]
factor_dict = dict(zip(LFPM_types, factor_sets))
alphas = dict(zip(LFPM_types, [[], [], []]))
params = {fm: [] for fm in LFPM_types}
# 1.2.1 CAPM


for ticker in tickers:
    y = excess_stocks_rets[ticker]

    for fm in LFPM_types:
        factor_set = factor_dict[fm]
        X = sm.add_constant(factor_rets[factor_set])
        model = sm.OLS(y, X).fit()

        alpha = model.params['const']
        param = model.params.to_numpy()
        alphas[fm].append(alpha)
        params[fm].append(param)

for factor_model in LFPM_types:
    annualized_abs_alpha[factor_model] = [FREQ * sum([abs(alpha) for alpha in alphas[factor_model]])/len(alphas[factor_model])]

df2 = pd.DataFrame(annualized_abs_alpha)
display(df2)


```

```python
import matplotlib.pyplot as plt

actual_values = annualized_means
for fm in LFPM_types:
    predicted_values = []
    factor_set = factor_dict[fm]
    factor_expected_rets = [factor_rets[factor].mean() for factor in factor_set]
    for ticker, param in zip(tickers, params[fm]):
        fm_betas = param[1:]
        predicted_values.append(fm_betas @ factor_expected_rets)

    plt.scatter(predicted_values, actual_values)
    plt.xlabel('model risk premia')
    plt.ylabel('sample risk premia')
    plt.title(fm)

    plt.show()

```
